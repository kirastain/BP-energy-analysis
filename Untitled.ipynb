## 4. Анализ данных и применение метода K means
Для начала получим суммарное значение по потреблению видов топлива, с которым и будем работать далее.

```python
data_2019['Total'] = data_2019['coalcons_mtoe'] + data_2019['gascons_mtoe'] + data_2019['oilcons_mtoe']
data_2019
```

```python
np.set_printoptions(suppress=True)
```

```python
array = np.array(data_2019['Total'])
```

Для определения нежного нам количества кластеров воспользуемся методом локтя. Для этого просмотрим все варианты количества кластеров на промежутке и по построенному графику SSE выявим наиболее оптимальное количество.

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42
}

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(array.reshape(-1, 1))
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

Для большей наглядности проверим выбранное нами значение, из графика равное 2.

```python
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

Примением метод K means

```python
kmeans = KMeans(n_clusters=kl.elbow, **kmeans_kwargs)
kmeans.fit(array.reshape(-1, 1))
```

```python
kmeans.labels_
```

Сделаем +1 для всех значений, чтобы далее отличать от стран, по которым нет данных

```python
kmeans.labels_ += 1
```

```python
data_2019['labels'] = kmeans.labels_
```

```python
data_2019[data_2019['labels'] > 1]
```

Из результатов применения метода получается неравномерный, но довольно очевидный результат: две страны, США и Китай, несравнимо впереди по потреблению углеводородов.
