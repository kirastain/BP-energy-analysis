```python
import pandas as pd
import os
import math
import geopandas as gpd
import matplotlib.pyplot as plt
import numpy as np
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
data = pd.read_csv("bp-stats-review-2020-consolidated-dataset-panel-format.csv")
```

```python
data.head()
```

```python
data_latest = data[data["Year"]==data["Year"].max()]
data_latest
```

```python
data.columns
```

```python
data_2019 = data_latest.loc[:,["Country", "Year", "Region", "coalcons_mtoe", "gascons_mtoe", "oilcons_mtoe"]]
data_2019 = data_2019.reset_index(drop=True)
data_2019
```

```python
data_2019[data_2019.iloc[:, 4:].isna().any(axis=1)]
```

```python
data_2019 = data_2019.drop(data_2019[data_2019.iloc[:, 4:].isna().any(axis=1)].index)
data_2019
```

```python
print(data_2019.Country.values)
```

```python
regions = [
    'Total Africa',
    'Total Asia Pacific',
    'Total CIS',
    'Total Central America',
    'Total Eastern Africa',
    'Total Europe',
    'Total European Union',
    'Total Middle Africa',
    'Total Middle East',
    'Total Non-OECD',
    'Total North America',
    'Total OECD',
    'Total S. & Cent. America',
    'Total Western Africa',
    'Total World'
]
```

```python
for part in regions:
    index_delete = data_2019[data_2019['Country'] == part].index
    data_2019.drop(index_delete, inplace=True)
index_delete = data_2019[data_2019['Country'] == 'Singapore'].index
data_2019.drop(index_delete, inplace=True)
index_delete = data_2019[data_2019['Country'] == 'China Hong Kong SAR'].index
data_2019.drop(index_delete, inplace=True)
data_2019 = data_2019.replace('North Macedonia', 'Macedonia')
data_2019 = data_2019.replace('US', 'United States of America')
```

```python
data_2019['Total'] = data_2019['coalcons_mtoe'] + data_2019['gascons_mtoe'] + data_2019['oilcons_mtoe']
data_2019
```

```python
np.set_printoptions(suppress=True)
```

```python
array = np.array(data_2019['Total'])
```

```python
kmeans_kwargs = {
    "init": "random",
    "n_init": 10,
    "max_iter": 300,
    "random_state": 42
}

sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, **kmeans_kwargs)
    kmeans.fit(array.reshape(-1, 1))
    sse.append(kmeans.inertia_)
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(range(1, 11), sse, curve="convex", direction="decreasing")
kl.elbow
```

```python
kmeans = KMeans(n_clusters=kl.elbow, **kmeans_kwargs)
kmeans.fit(array.reshape(-1, 1))
```

```python
kmeans.labels_
```

```python
kmeans.labels_ += 1
```

```python
data_2019['labels'] = kmeans.labels_
```
